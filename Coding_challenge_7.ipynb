Built-in Functions and User-Defined Functions
Project Title: Simple Calculator App

Scenario:
 A company wants to provide users with a simple calculator to perform basic mathematical operations.

Tasks:
Create functions for addition, subtraction, multiplication, and division.


Ask the user to input two numbers and the operation they want to perform.


Call the appropriate function and display the result.


Use built-in functions like abs(), round(), and pow() within your solution.
Handle division by zero gracefully and inform the user.


```python
# Simple Calculator Application

# User defined function
def add(x,y):
  return x+y


def sub(x,y):
 return x-y


def multiply(x,y):
  return x*y


def divide(x,y):
  if y==0:
    print("division by zero is not possible")
    return "Error: Division by zero"
  else:
    return x/y

def power(x,y):
  return pow(x,y)

print("=====Simple Calculator=======")
num1=float(input("Enter first number: "))
num2=float(input("Enter second number: "))

print("\nSelect operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")

choice=int(input("Enter choice(1/2/3/4/5): "))
result = None
if choice==1:
  result=add(num1,num2)
  print("Result: ",result)
elif choice==2:
  result=sub(num1,num2)
  print("Result: ",result)
elif choice==3:
  result=multiply(num1,num2)
  print("Result: ",result)
elif choice==4:
  result=divide(num1,num2)
  # The divide function prints an error message, so we only print result if it's a number
  if isinstance(result, (int, float)):
    print("Result: ",result)
  else:
    print(result)
elif choice==5:
  result=power(num1,num2)
  print("Result: ",result)
else:
  print("Invalid choice")

# Using built-in functions
# Only apply built-in functions if result is a number
if isinstance(result, (int, float)):
    print("\nResult:", round(result, 2))
    print("Absolute value:", abs(result))
else:
    print("\n", result)

```
