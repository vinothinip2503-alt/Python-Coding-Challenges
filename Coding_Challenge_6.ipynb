```python
item_count = 0
subtotal = 0.0

while True:
    try:
        price = float(input("Enter item price : "))

        # Sentinel value to stop
        if price == 0:
            break

        # Check for negative price
        if price < 0:
            print("Negative prices are not allowed. Please enter again.")
            continue

        # Valid price
        item_count += 1
        subtotal += price

    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Apply discount based on subtotal
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Display bill summary
print("\n----- BILL SUMMARY -----")
print(f"Total Items: {item_count}")
print(f"Subtotal: ₹{subtotal:.2f}")
print(f"Discount: ₹{discount_amount:.2f} ({int(discount_rate*100)}%)")
print(f"Final Total: ₹{final_total:.2f}")
print("------------------------")
```
